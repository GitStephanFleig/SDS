```python
# Loading packages

import requests
from bs4 import BeautifulSoup
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import re
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load df from disk

df = pd.read_csv("speeches/data.csv")
```

```python
df.loc[134]['Speech']
```

```python
df['Speech_Parsed_1'] = df['Speech'].str.replace("\r", " ")
df['Speech_Parsed_1'] = df['Speech_Parsed_1'].str.replace("\n", " ")
df['Speech_Parsed_1'] = df['Speech_Parsed_1'].str.replace("    ", " ")
df['Speech_Parsed_1'] = df['Speech_Parsed_1'].str.replace('"', '')
```

```python
# Lowercasing the text
df['Speech_Parsed_2'] = df['Speech_Parsed_1'].str.lower()
```

```python
# Punctuation signs
punctuation_signs = list("?:!.,;")
df['Speech_Parsed_3'] = df['Speech_Parsed_2']

for punct_sign in punctuation_signs:
    df['Speech_Parsed_3'] = df['Speech_Parsed_3'].str.replace(punct_sign, '')
```

```python
df.loc[134]['Speech_Parsed_3']
```

```python
# Split speech by " " into words
df['Speech_words']= df['Speech_Parsed_3'].str.split(" ")
```

```python
# calculating mean word acount and adding to df

count=[]
wordCount=0
letterCount=0

for i in range(0,len(df)): 
    for word in df.loc[i]['Speech_words']:
        import time
        # noz counting empty words:
        if len(word) > 0:
            letterCount=letterCount+len(word)
            wordCount=wordCount+1
    mean = letterCount/wordCount
    count.append(mean)
    
df['mean_word_length'] = count
```
